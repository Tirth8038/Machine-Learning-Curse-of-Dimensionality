```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.spatial.distance import cdist
```

```python
#Creating a function to Generate the Data based on Number of Samples and Dimensions.
def gen_data(sample,dim):
    class_1=np.random.normal(0,1,(sample,dim))
    class_2=np.random.normal(3,1,(sample,dim))
    data=np.concatenate([class_1,class_2], axis=0)
    data=(data-data.mean(axis=0))/data.std(axis=0)
    
    return data,cdist(data,data,metric="euclidean").flatten()
```

```python
#Get 2D Data
sample_number=20
data,distances=gen_data(sample_number,2)
print(data[:5])
plt.scatter(data[:sample_number,0], data[:sample_number,1], s=10, color='r')
plt.scatter(data[sample_number:,0], data[sample_number:,1], s=10, color='b')
plt.show()
sns.distplot(distances, hist=True, kde=True, color='g')
plt.show()

```

```python
#Higher Dimentional Space
#As the Dimension Space increases the Data becomes more Sparse and the Histogram becomes less densed and far apart from each other.
for dim in list(range(3,)) + list(range(10,110,50)) + [100*i for i in range(2,6)]:
    _, distances=gen_data(sample_number,dim)
    sns.distplot(distances, hist=True, kde=True, color='g')
    plt.title("Dimension = %s"%dim)
    plt.show()
```
